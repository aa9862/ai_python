```python
# 출력(반환값):함수의 작업 결과물

def mood_check(score):
  if score >=8:
    return "기분이 아주 좋아보이네요 ! 즐거운 하루 되세요."
  elif score >=5 :
    return "그럭저럭 괜찮은 하루군요. 힘내세요."
  else :
    return "조금 우울하신가요? 따뜻한 차 한잔 어때요?"

user_score = int(input("오늘 기분 점수는? (0~10): "))
result_message = mood_check(user_score)
print(result_message)
```

```python
#실생활 예제로 이해하기

def coffee_order(menu) :
  if menu == "아메리카노":
    return "진한 에스프레소에 시원한 물 한 잔 추가했어요!"
  elif menu == "라떼":
    return "부드러운 우유 거품과 함께 나갑니다."
  elif menu == "바닐라라떼":
    return "달콤한 바닐라 향으로 기분 좋은 하루 시작하세요!"
  else :
    return "죄송해요, 해당 메뉴는 준비되지 않았어요."

drink = input("무슨 커피를 주문하시겠어요? ")
response = coffee_order(drink)
print(response)
```

```python
def bmi_fun(h,w):
  bmi = w/h**2
  return bmi
```

```python
h= float(input("키 : "))
w= float(input("몸무게 :"))
bmi = round(bmi_fun(h,w),3)
print(bmi)
```

```python
celsius =[0, 10, 20, 30, 40]
fahrenheit = list(map(lambda x: (x*9/5)+32, celsius))
print(f"섭씨 : {celsius}")
print(f"화씨:{fahrenheit}")
```
