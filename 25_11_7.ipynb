```python
import numpy as np
import time

# 1. 파이썬 리스트를 사용한 경우
my_list = list(range(1_000_000))
start_time = time.time()
result_list = [x * 2 for x in my_list]
end_time = time.time()
print(f"파이썬 리스트 실행 시간: {end_time - start_time:.5f}초")


# 2. 넘파이 배열을 사용한 경우
my_array = np.arange(1_000_000)
start_time = time.time()
result_array = my_array * 2
end_time = time.time()
print(f"넘파이 배열 실행 시간: {end_time - start_time:.5f}초")
```

```python
import numpy as np
```

```python
distances = np.array([1,2,3,5,8])

wind_speeds = np.array([0,5,10,7,12])
```

```python
base_usage_rate = 10

additional_usafe_rate = wind_speeds
```

```python
total_usage_rate = base_usage_rate + additional_usafe_rate
```

```python
battery_consumption = distances * total_usage_rate
```

```python
print("비행거리(km) : ", distances)
print("풍속(m/s) : ", wind_speeds)
print("총 소모율(%/km) : ", total_usage_rate)
print("예상 배터리 소모량(%)", battery_consumption)
```
